```python
import numpy as np
with open('input.txt', 'r') as file:
    data = file.read()
    data = data.split("\n\n")
    presents = {}
    areas = []
    
    for _ in range(len(data)-1):
        line = data.pop(0)
        if "#" in line:
            id = int(line.split(":")[0])
            shape = line.split(":")[1].strip().split("\n")
            shape_array = []
            for row in shape:
                shape_array.append([1 if char == "#" else 0 for char in row])
            shape = np.array(shape_array)
            presents[id] = shape
            continue
    data = data[0].split("\n")
    for line in data:
        if not line:
            continue
        dim = line.split(":")[0]
        if "x" in dim:
            dim_values = dim.split("x")
            shape = (int(dim_values[0]), int(dim_values[1]))
        n_presents = line.split(":")[1].split(" ")

        areas.append((shape, [int(n) for n in n_presents if n.isdigit()]))
    
```

## Part 1

```python
fit = 0
for area in areas:
    shape = area[0]
    n_presents = area[1]
    sheet = np.zeros(shape)
    available_squares = np.prod(shape)
    required_squares = 0
    for i, n in enumerate(n_presents):
        present_shape = presents[i]
        required_squares += (np.sum(present_shape) * n)
    if required_squares > available_squares:
        # Not enough space to begin with
        continue
    all_fit = []
    for i, n in enumerate(n_presents):
        present_shape = presents[i]
        present_area = np.sum(present_shape)
        placed = False
        for mirror in range(2):
            if mirror == 1:
                present_shape = np.fliplr(present_shape)
            for rotation in range(4):
                present_shape = np.rot90(present_shape)
                for _ in range(n):
                    
                    for row in range(shape[0] - present_shape.shape[0] + 1):
                        if placed:
                            break
                        for col in range(shape[1] - present_shape.shape[1] + 1):
                            sub_area = sheet[row:row+present_shape.shape[0], col:col+present_shape.shape[1]]
                            if np.sum(sub_area + present_shape) == present_area:
                                sheet[row:row+present_shape.shape[0], col:col+present_shape.shape[1]] += present_shape
                                placed = True
                                break
                    if not placed:
                        break
                else:
                    continue
                break
        if placed:
            all_fit.append(True)
    if len(all_fit) == len(n_presents):
        fit += 1
```

```python
with open("input.txt", "r") as file:
    data = file.read()
    data = data.splitlines()
```

## Part 1

```python
# Parse ranges and ids from input data
def ranges_and_ids(data):
    ranges = []
    ids = []
    for line in data:
        if not line:
            continue
        if "-" not in line:
            ids.append(int(line))
        else:
            start_str, end_str = line.split("-")
            start, end = int(start_str), int(end_str)
            ranges.append((start, end))
    return ranges, ids
ranges, ids = ranges_and_ids(data)
```

```python
# concatenate ranges if overlapping
def concatenate_ranges(ranges):
    if not ranges:
        return []
    ranges.sort()
    concatenated = [ranges[0]] # initialize with the first range
    for current in ranges[1:]:
        last = concatenated[-1] # get the last range in concatenated
        if current[0] <= last[1] + 1:
            concatenated[-1] = (last[0], max(last[1], current[1])) # merge ranges
        else:
            concatenated.append(current) # add new range#
    return concatenated
concatenated_ranges = concatenate_ranges(ranges)
    
```

```python
# Find ids that are contained in any of the ranges
def find_ids_in_ranges(ids, ranges):
    ids_in_ranges = set()
    for start, end in ranges:
        for id_ in ids:
            if start <= id_ <= end:
                ids_in_ranges.add(id_)
    return ids_in_ranges

print("# of contained ids:", len(find_ids_in_ranges(ids, concatenated_ranges)))
```

## Part 2

```python
# now all ids in ranges are valid
sum_of_ranges = sum(end - start + 1 for start, end in concatenated_ranges)
print("Total ids in ranges:", sum_of_ranges)
```

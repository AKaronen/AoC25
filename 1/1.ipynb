```python
with open('input.txt', 'r') as file:
    data = file.read()
    data = data.splitlines()
    data = [(x[0], int(x[1:])) for x in data]
```

## Part 1

```python
import numpy as np
ops = {"R": np.add, "L": np.subtract}
count_zeros = 0
cur_pos = 50
for i, (direction, value) in enumerate(data):
    op = ops[direction] # choose operation (R to add, L to subtract)
    cur_pos = op(cur_pos, value) % 100 # update position modulo 100 (0-99)
    #print(f"Step {i}: {direction}{value} -> {cur_pos}")
    if cur_pos == 0:
        count_zeros += 1

print(count_zeros)
    
```

## Part 2

```python
ops = {"R": np.add, "L": np.subtract}
count_zeros = 0
cur_pos = 50
for i, (direction, value) in enumerate(data):
    op = ops[direction]  # choose operation
    for _ in range(value): # step through each unit of movement at a time and check for zeros
        cur_pos = op(cur_pos, 1) % 100
        if cur_pos == 0:
            count_zeros += 1
print(count_zeros)
```

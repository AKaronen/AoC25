```python
with open("input.txt", "r") as file:
    data = file.read()
    data = data.splitlines()
```

## Part 1

```python
total_joltage = 0
for bank in data:
    digits = [int(d) for d in bank]
    idx, max_digit = max(enumerate(digits), key=lambda x: x[1]) # get index and value of max digit
    if idx != len(digits) - 1: # max digit is not the last digit -> take from the back
        rest = digits[idx + 1 :]
        max_from_rest = max(rest)    
        joltage = int(str(max_digit) + str(max_from_rest))
    else: # max digit is the last digit -> take from the front
        rest = digits[:idx]
        max_from_rest = max(rest)
        joltage = int(str(max_from_rest) + str(max_digit))

    total_joltage += joltage
print(f"Total joltage: {total_joltage}")
    
    
```

## Part 2

```python
def max_subdigit(arr, length):
    """ Returns the maximum subdigit of given length from the array of digits 
    (digits don't have to be consecutive but have to read from left to right).
    """
    max_digit = -1
    n = len(arr)
    subdigit = [0] * length
    max_val_idx = lambda x: max(enumerate(x), key=lambda y: y[1])  # noqa: E731
    idx , init_max_val = max_val_idx(arr[: n - length + 1]) # first digit candidate (if multiple, max returns first :D)
    subdigit[0] = init_max_val 
    last_idx = idx
    for i in range(1, length):
        idx_offset = last_idx + 1 # next digit must be after the last chosen digit
        idx, val = max_val_idx(arr[idx_offset : n - length + i + 1]) # next digit candidate
        subdigit[i] = val 
        last_idx = idx + idx_offset # adjust index to original array
    max_digit = int(''.join(str(d) for d in subdigit))
    return max_digit

total_joltage = 0
for bank in data:
    digits = [int(d) for d in bank]
    joltage = max_subdigit(digits, 12)
    total_joltage += joltage
print(f"Total joltage: {total_joltage}")


    
```

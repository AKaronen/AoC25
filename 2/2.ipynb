```python
with open('input.txt', 'r') as file:
    data = file.read()
    data = data.split(',')
    data = [(int(x.split('-')[0]), int(x.split('-')[1])) for x in data] # Parse input into list of tuples (ranges)
```

## Part 1

```python
invalid_accumulator = 0
for r in data:
    start, end = r
    for number in range(start, end + 1):
        number = str(number)
        if number[0:len(number)//2] == number[len(number)//2:]:
            invalid_accumulator += int(number)

print(f"Total invalid sum: {invalid_accumulator}")
     
```

## Part 2

```python
import math
def divisors(n):
    divs = [1]
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            divs.extend([i, n / i])
    divs.extend([n])
    return list(set(divs))
```

```python
invalid_accumulator = 0
for r in data:
    start, end = r
    for number in range(start, end + 1):
        number = str(number)
        seq_len = len(number)
        divs = divisors(seq_len) # get all possible subsequence lengths
        for subseq_len in divs:
            subseq_len = int(subseq_len)
            if subseq_len == seq_len:
                continue
            subseq = number[0:subseq_len] # get the candidate subsequence
            if all(subseq == number[i:i+subseq_len] for i in range(0, seq_len, subseq_len)): # check if the entire number is made up of this subsequence
                invalid_accumulator += int(number)
                break

print(f"Total invalid sum: {invalid_accumulator}")
```

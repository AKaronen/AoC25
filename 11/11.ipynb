```python
example_data = [
    "aaa: you hhh",
"you: bbb ccc",
"bbb: ddd eee",
"ccc: ddd eee fff",
"ddd: ggg",
"eee: out",
"fff: out",
"ggg: out",
"hhh: ccc fff iii",
"iii: out"
]
with open("input.txt", "r") as f:
    data = f.read().splitlines()
    
```

```python
def parse_data(data):
    nodes = {}
    
    for line in data:
        target, rest = line.split(": ")
        parts = rest.split(" ")
        
        nodes[target] = parts
    return nodes
data_parsed = parse_data(data)
```

## Part 1

```python
num_paths = 0
def count_paths(node):
    global num_paths
    global data_parsed
    if node == "out":
        num_paths += 1
        return
    for neighbor in data_parsed[node]:
        count_paths(neighbor)

count_paths("you")
print(num_paths)
```

```python
example_data = [
    "svr: aaa bbb",
"aaa: fft",
"fft: ccc",
"bbb: tty",
"tty: ccc",
"ccc: ddd eee",
"ddd: hub",
"hub: fff",
"eee: dac",
"dac: fff",
"fff: ggg hhh",
"ggg: out",
"hhh: out",
]

data_parsed = parse_data(data)
```

```python
# modify into a connection matrix
import numpy as np
size = len(data_parsed)
C = np.zeros((size+1, size+1), dtype=int) # add and extra row/col for "out"
index_map = {key: i for i, key in enumerate(data_parsed.keys())}
for key, neighbors in data_parsed.items():
    i = index_map[key]
    for neighbor in neighbors:
        if neighbor == "out":
            j = size  # index for "out"
        else:
            j = index_map[neighbor]
        C[i, j] = 1
C = C.astype(np.int64)
powers_of_C = np.array([np.linalg.matrix_power(C, k) for k in range(1, 20)]) # assuming max path length < 20
```

```python
svr_index = index_map["svr"] # Get index of "svr"
fft_index = index_map["fft"] # Get index of "fft"
dac_index = index_map["dac"] # Get index of "dac"
out_index = size  # index for "out"
# Calculate total paths going through fft and dac (svr->fft->dac->out)
total_paths = sum(powers_of_C[:, svr_index, fft_index]) * sum(powers_of_C[:, fft_index, dac_index]) * sum(powers_of_C[:, dac_index, out_index])
print(total_paths)
```
